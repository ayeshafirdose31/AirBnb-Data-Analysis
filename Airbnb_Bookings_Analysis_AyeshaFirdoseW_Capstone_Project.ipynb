# **Outliers**: An outlier may indicate bad data, so we will visualize an check.

```python
#chcking the outliers wih respect to pries in our dataset
fig, ax = plt.subplots(figsize=(8, 8))
sns.boxplot(data_AirBnb['price'])
```

```python
 #Removing Outliers
 Q1 = np.percentile(data_AirBnb['price'],25)
 Q2 = np.percentile(data_AirBnb['price'],50)
 Q3 = np.percentile(data_AirBnb['price'],75)
 IQR =  Q3 - Q1
 a=Q1-(1.5*IQR)
 b=Q3+(1.5*IQR)
 data_AirBnb=data_AirBnb[(data_AirBnb['price']<b) & (data_AirBnb['price']>a)]
```

# **4) Is there any noticeable difference of traffic among different areas and what could be the reason for it?**
